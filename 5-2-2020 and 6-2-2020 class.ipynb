```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("Advertising.csv")
data.tail()
```

```python
data.drop("Unnamed: 0",axis=1,inplace=True)
data.tail()
```

```python
x=data["TV"].values
print(x.shape)
print(x)
```

```python
y=data["sales"]
print(y.shape)
y
```

```python
x_mean=x.mean()
print("x-mean:",x_mean)
y_mean=np.mean(y)
print("y-mean:",y_mean)
```

```python
# calculating error
n=len(x)
print(n)   #y=b1x+b0   b1-->slope,b0-->intercept
numer=0   
denom=0
for i in range(n):
    numer+=(x[i]-x_mean)*(y[i]-y_mean)
    denom+=(x[i]-x_mean)**2
b1=numer/denom
b0=y_mean-(b1*x_mean)
print("slope:",b1)
print("intercept:",b0)
```

```python
#y=mx+c
print("linear equation:y=[:.5]x+[:.4]".format(b1,b0))
```

```python
y_predict=b1*x+b0
y_predict
```

```python
# r2 score
n1=0
d1=0
for i in range(n):
    n1+=(y_predict[i]-y_mean)**2
    d1+=(y[i]-y_mean)**2
r2=1-n1/d1
print("r2 score:",r2)
```

```python
plt.scatter(x,y_predict)
```

```python
plt.scatter(x,y_predict,label="linear wave",color="g")
plt.xlabel("x")
plt.ylabel("y_predict value")
plt.title("linear regression with model including")
plt.legend()
```
